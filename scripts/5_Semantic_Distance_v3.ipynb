```python
# Generate sentence 1 - EN native
doc1 = nlp("For many States, the lack of sufficient capacity represents a major challenge in effectively monitoring and enforcing the arms embargo.")
# Generate sentence 2 - AR EN DeepL
doc2 = nlp("For many states, lack of sufficient capacity is a major challenge in monitoring and effectively enforcing the arms embargo.")
# Generate sentence 3 - AR EN Deep
doc3 = nlp("The lack of sufficient capacity is a major challenge for many States in the effective monitoring and enforcement of arms embargoes.")
```
